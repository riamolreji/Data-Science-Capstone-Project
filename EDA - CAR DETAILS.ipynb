### Q8) How many cars have high, medium and low selling_price w.r.t total average selling_price?

```python
# Calculate the average (mean) selling price
average_selling_price = df['selling_price'].mean()
```

```python

# Define the thresholds based on the average selling price
high_threshold = 2 * average_selling_price  # Above twice the average is High Budget Cars
low_threshold = average_selling_price / 2  # Below half the average is Low Budget Cars

# Create a new column 'category' to label the cars
def categorize_car(selling_price):
    if selling_price >= high_threshold:
        return 'High'
    elif selling_price <= low_threshold:
        return 'Low'
    else:
        return 'Medium'

df['category'] = df['selling_price'].apply(categorize_car)

# Split the dataset into three based on the category
dfh = df[df['category'] == 'High']
dfm = df[df['category'] == 'Medium']
dfl = df[df['category'] == 'Low']

# Print the number of cars in each category
print("Number of High Budget Cars:", len(dfh))
print("Number of Medium Budget Cars:", len(dfm))
print("Number of Low Budget Cars:", len(dfl))
```
